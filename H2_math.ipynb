```python
from geomstats.geometry.hyperbolic import Hyperbolic
import geomstats.visualization as viz
from geomstats.learning.frechet_mean import FrechetMean


import numpy as np
import matplotlib.pyplot as plt
```

```python
h2 = Hyperbolic(
    2, default_coords_type = 'half-space'
)
```

```python
toplt = h2.random_point(n_samples = 10)
```

```python
viz.plot(toplt, space = 'H2_poincare_half_plane')
```

```python
xmean = np.mean(toplt[:,0])
ymean = np.mean(toplt[:,1])
r2mean = np.array([xmean, ymean])
```

```python
mean = FrechetMean(metric = h2.metric)
mean.fit(toplt)

ptmean = mean.estimate_

fig = plt.figure()
ax = viz.plot(toplt, space = 'H2_poincare_half_plane')
ax = viz.plot(ptmean, space='H2_poincare_half_plane', marker='*', color='red', s=200)
ax = viz.plot(r2mean, space='H2_poincare_half_plane', marker='*', color='green', s=200)


```
